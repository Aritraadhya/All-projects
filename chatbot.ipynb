```python
#data extraction from this file
#from pytesseract
import re
import PyPDF2

def extract_text_from_pdf(pdf_path):
    try:
        with open(pdf_path, 'rb') as file:
            pdf_reader = PyPDF2.PdfReader(file)
            text = ''
            for page_num in range(len(pdf_reader.pages)):
                page = pdf_reader.pages[page_num]
                text += page.extract_text() + '\n'
            return text
    except PyPDF2.errors.PdfReadError as e:
        return f"Error reading PDF: {e}"
    except Exception as e:
        return f"An error occurred: {e}"

def extract_data_with_questions(text, questions):
    extracted_data = {}
    for question in questions:
        # Create a pattern to match potential answers following the question
        pattern = re.compile(rf'{re.escape(question)}\s*:\s*(.*)', re.IGNORECASE)
        match = pattern.search(text)
        if match:
            extracted_data[question] = match.group(1).strip()
        else:
            extracted_data[question] = 'Not found'
    return extracted_data
# Example usage
pdf_path = "C:\\Users\\Admin\\Downloads\\My_Resume_(16).pdf"  # Replace with your PDF file path
#prompts = ['Klasse', 'Erw', 'Hinfahrt', 'Gültig ab','Auftragsnummer','Über: VIA']  # Replace with your prompts
questions = [
    'Klasse',
    'Erw',
    'Gültig ab',
    'Auftragsnummer',
    'Über'
]

# Extract text from the PDF
extracted_text = extract_text_from_pdf(pdf_path)
print(extracted_text)

#if "An error occurred" not in extracted_text:
    # Extract data based on questions
 #   extracted_data = extract_data_with_questions(extracted_text, questions)
  #  for question, answer in extracted_data.items():
   #     print(f'{question} {answer}')
#else:
 #   print(extracted_text)
```

```python
#convert all the above information in JSON format
import json

def text_to_json(text):
    # Split the text into lines
    lines = text.splitlines()

    # Create a dictionary to hold the key-value pairs
    data = {}
    current_key = None
    current_value = []

    # Iterate through each line
    for line in lines:
        # Check if the line contains a key-value pair (with a ":")
        if ":" in line:
            # If there's a current key, save its value before moving to the next key
            if current_key:
                data[current_key] = "\n".join(current_value).strip()
            
            # Split the line into key and the beginning of the value
            key, value = line.split(":", 1)
            current_key = key.strip()
            current_value = [value.strip()]
        else:
            # If there's no ":", treat this line as a continuation of the current value
            current_value.append(line.strip())

    # After the loop, save the last key-value pair
    if current_key:
        data[current_key] = "\n".join(current_value).strip()

    # Convert the dictionary to a JSON string
    json_data = json.dumps(data, indent=4)

    return json_data

json_output = text_to_json(response)
print(json_output)
```
